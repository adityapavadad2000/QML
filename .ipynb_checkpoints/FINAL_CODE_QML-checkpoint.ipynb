<div align="center" style="font-size:24px;"><b>CODE FOR VEHICLE ROUTING PROBLEM</b></div>


```python
# ---- IMPORT LIBRARIES ----

import matplotlib.pyplot as plt
from gurobipy import *
import numpy as np
import csv

#---- IMPORTING DATASET AND EXTRACTING THE APPRORIATE DATA ----

filename = "data_large_3dep.txt"


with open(filename) as f:
    data = csv.reader(f, delimiter="\t")
    dataset = list(data)

for i in range(0,len(dataset)):
    for j in range(0,7):
        a = int(dataset[i][j])
        dataset[i][j] = a
 
#---- PARAMETERS ----

dataset = np.array(dataset)
node_id = dataset[:, 0]  # Stores the id's of the nodes
x_coord = dataset[:, 1]  # Stores the x coordinates of all the nodes
y_coord = dataset[:, 2]  # Stores the y coordinates of all the nodes
a = dataset[:, 3]  # Stores the demand for each node
RT = dataset[:, 4]  # Stores the ready time for each node (earliest time for delivery)
DT = dataset[:, 5]  # Stores the due time for each node (latest time for delivery)
ST = dataset[:, 6]  # Stores the service time for each node (time needed for a delivery at each location )





# ---- CAPACITIES OF EACH VEHICLE ----

b = np.array([300 ,300 ,300 ,300 ,300,300]) 

# ---- COST PER DEPOT ----

Dc = np.array([5, 0, 0])


#---- TOTAL NUMBER OF VEHICLES ----

K = len(b)

#---- CALCULATING THE EUCLIDEAN DISTANCE BETWEEN ALL NODES ----

D = np.zeros((len(a), len(a)))


for i in range(len(node_id)):
    for j in range(len(node_id)):
        Distance = np.linalg.norm((x_coord[i] - x_coord[j], y_coord[i] - y_coord[j]))
        D[i , j] = round(Distance , 2)
        

# Note for this programme, the assumption is made that 1 distance unit corresponds to 1 time unit
# This also allows us to use distance interchangably within constraints with time


# ---- INITIALIZING  MODEL ----

model = Model("Part a/e/h")


# ---- SETS AND INDICES ----

g=3  #Number of Depots
I = range(len(node_id))  # locations
J = range(len(node_id))  # locations
depots = range(g) #Set for depots
customers = range(g, len(node_id)) #Set for customers
V = range(K)  # Vehicles

# ---- DECISION VARIABLES ----

x = model.addVars(I, J, V, vtype = GRB.BINARY, name = "If_Route_Travelled")
t = model.addVars(I, V, vtype = GRB.CONTINUOUS, name = "Service_Start_Time")
    
# ---- INTEGERATE THE VARIABLES INTO THE MODEL ----

model.update()

# ---- OBJECTIVE FUCNTION ----

objective = (quicksum(D[i, j] * x[i, j, v] for i in I for j in J for v in V if i != j) +
             quicksum(Dc[d] * quicksum(x[d, j, v] for j in customers) for d in depots for v in V))
# objective = quicksum(D[i, j] * x[i, j, v] for i in I for j in J for v in V if i != j) 

model.setObjective(objective, GRB.MINIMIZE)


model.update()

# ---- INITIALIZING THE BIG M VALUE ----

M = 3000

# ---- CONSTRAINTS ----

# Constraint 1: Each customer is visited exactly once

single_visit = model.addConstrs(
    (quicksum(x[i, j, v] for j in J for v in V if j != i) == 1 for i in customers),
    "Single_Visit_Satisfaction")


# Constraint 2: Not exceeding the total number of vehicles

depot_vehicle_num = model.addConstr(
    (quicksum(x[d, j, v] for j in customers for v in V for d in depots) <= K),
    "Vehicle_Leaving_Depot")


# Constraint 3: Vehicle capacity is not exceeded

vehicle_cap = model.addConstrs(
    (quicksum(a[j] * x[i, j, v] for j in customers for i in I if i != j) <= b[v] for v in V ),
    name="Vehicle_Capacity")

# Constraint 4: Each vehicle returns to the same depot it left from

return_to_same_depot = model.addConstrs(
    (quicksum(x[i, j, v] for j in customers) == quicksum(x[j, i, v] for j in customers) 
     for i in depots for v in V),
    "Return_To_Same_Depot")


# Constraint 5: Each vehicle must leave from exactly one depot

unique_depot_start = model.addConstrs(
    (quicksum(x[d, j, v] for d in depots for j in customers) == 1 for v in V),
    "Unique_Depot_Start")

# Constraint 6: Each vehicle must return to a depot

vehicle_return = model.addConstrs(
    (quicksum(x[j, d, v] for j in customers if j != d) == quicksum(x[d, j, v] for j in customers if j != d) for d in depots for v in V),
    "Vehicle_Return_To_Original_Depot")


# Constraint 7: Route continuity (flow conservation)

route_flow_customers = model.addConstrs(
    (quicksum(x[i, j, v] for j in customers if i != j) +
     quicksum(x[i, d, v] for d in depots) - 
     quicksum(x[j, i, v] for j in customers if i != j) -
     quicksum(x[d, i, v] for d in depots) == 0 for i in customers for v in V),
    "Maintaining_Route_Flow_Customers")


route_flow_depots = model.addConstrs(
    (quicksum(x[d, j, v] for j in customers) - 
     quicksum(x[j, d, v] for j in customers) == 0 for d in depots for v in V),
    "Maintaining_Route_Flow_Depots")



# Constraint 8: Service start time is within the ready time and due time

service_start_time = model.addConstrs(
    (t[i, v] >= RT[i] for i in customers for v in V),
    "Service_Start_Not_Before_Ready_Time")

service_end_time = model.addConstrs(
    (t[i, v] <= DT[i] for i in I for v in V),
    "Service_End_Before_Due_Time")

# Constraint 9: No self return

no_self_return_customers = model.addConstrs(
    (x[i, i, v] == 0 for i in customers for v in V),
    "No_Self_Return_Customers")

no_self_return_depots = model.addConstrs(
    (x[d, d, v] == 0 for d in depots for v in V),
    "No_Self_Return_Depots")


# Constraint 10: Service time window with travel time and service time consideration

customers_list = list(customers)
depots_list = list(depots)

service_time_window_extended = model.addConstrs(
    (t[j, v] >= t[i, v] + ST[i] + D[i, j] - M * (1 - x[i, j, v])
     for i in customers_list + depots_list for j in customers_list for v in V if i != j),
    "Extended_Service_Time_Window")



# ---- ENABLE DETAILED OUTPUT LOGGING ----

model.setParam("OutputFlag", True)  # Set to True to enable logging

# ---- SET THE OPTIMALITY GAP (Eg.,0.05% or 0.0005 )

model.setParam("MIPGap", 0.0005)

# ---- SET A TIME LIMIT FOR THE SOLVER (IN SECONDS ,Eg., 1800 secs) ----

model.setParam("TimeLimit", 1800)

# model.setParam("Heuristics", 0.5)

# ----UPDATE THE MODEL TO APPLY THE PARAMETERS ----

model.update()

# ---- OPTIMIZE THE MODEL ----

model.optimize()

# After optimization, you can access the MIPGap

if model.SolCount > 0:
    mip_gap = model.MIPGap
    print("Optimality Gap:", mip_gap)
else:
    print("No solution found or solver did not reach optimality gap within time limit.")


```

<div align="center" style="font-size:24px;"><b>CODE FOR FINDING THE ROUTE</b></div>


```python
# Check if the model has a feasible solution
if model.status == GRB.OPTIMAL or model.status == GRB.TIME_LIMIT:
    
    # Print the routes for each vehicle
    for v in V:
        route_load = 0
        print(f"\nRoute for Vehicle {v}:")

        for i in customers:
            for j in customers:
                if i != j and x[i, j, v].X > 0.5:  # Check if edge is in the solution
                    route_load += a[i]  # Update load
                    print(f"    Node {i} to {j}, Load: {route_load}")
```

<div align="center" style="font-size:24px;"><b>CODE FOR PLOTTING THE ROUTE GRAPH</b></div>


```python
def get_route_for_vehicle(vehicle, x, depots, customers):
    # Find the starting depot for the vehicle
    start_depot = next(d for d in depots for j in customers if x[d, j, vehicle].X > 0.5)

    route = [start_depot]
    current_node = start_depot
    

    # Loop to find the route
    while True:
        next_node = next((j for j in customers if x[current_node, j, vehicle].X > 0.5), None)
        
        if next_node is None:  # Vehicle returns to a depot
            return_depot = next(d for d in depots if x[current_node, d, vehicle].X > 0.5)
            route.append(return_depot)
            break

        route.append(next_node)
        current_node = next_node

    return route
```

```python
if model.status == GRB.OPTIMAL or model.status == GRB.TIME_LIMIT:
    colors = plt.cm.tab10(np.linspace(0, 1, K))

    plt.figure(figsize=(12, 8))

    # Plot depots and customers with Node IDs
    plt.scatter(x_coord[list(depots)], y_coord[list(depots)], c='black', label='Depots', marker='D')
    plt.scatter(x_coord[list(customers)], y_coord[list(customers)], c='red', label='Customers', marker='o')

    # Offset for text annotations
    depot_offset = 1  # Adjust as necessary for your plot scale
    text_offset_x = 0.5  # Adjust as necessary for your plot scale
    text_offset_y = 0.5  # Adjust as necessary for your plot scale

    for i in range(len(node_id)):
        if i in depots:
            plt.text(x_coord[i] + depot_offset, y_coord[i] + depot_offset, str(node_id[i]), color='black', fontsize=8)
        else:
            plt.text(x_coord[i] + text_offset_x, y_coord[i] + text_offset_y, str(node_id[i]),
                 color='black' if i in depots else 'blue', fontsize=8)

    # Plot the routes for each vehicle
    for v in V:
        route = get_route_for_vehicle(v, x, depots, customers)
        route_color = colors[v]

        for i in range(len(route) - 1):
            start_x, start_y = x_coord[route[i]], y_coord[route[i]]
            end_x, end_y = x_coord[route[i + 1]], y_coord[route[i + 1]]
            
            dx, dy = end_x - start_x, end_y - start_y
            plt.arrow(start_x, start_y, dx, dy, color=route_color, length_includes_head=True,
                      head_width=1, head_length=0.7, overhang=1)

    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.title('Vehicle Routing with Multiple Depots and Node IDs')
    plt.legend()
    plt.grid()
    plt.show()
```

<div align="center" style="font-size:24px;"><b>CODE FOR FINDING THE LOAD AT EACH NODE </b></div>


```python
if model.SolCount > 0:
    # Print the routes and loads for each vehicle
    for v in V:
        print(f"\nVehicle {v} route with load drop-offs:")
        
        # Start with the depot node for vehicle v
        current_node = next((i for i in depots if any(x[i, j, v].X > 0.5 for j in customers)), None)
        
        # If no starting node is found for vehicle, there might be an issue with the model
        if current_node is None:
            print(f"    No route for vehicle {v}.")
            continue
        
        # Find the route for vehicle v
        route = [current_node]
        while True:
            next_node = next((j for j in I if j not in route and x[route[-1], j, v].X > 0.5), None)
            if next_node is None or next_node in depots:
                break
            route.append(next_node)

        # Calculate initial vehicle load as the sum of demands across the route
        vehicle_load = sum(a[j] for j in route if j in customers)
        print(f"    Initial load for Vehicle {v}: {vehicle_load}")

        # Go through the route and subtract demand at each node
        for node in route:
            if node in customers:  # Ignore depots since they don't have demand
                print(f"    Vehicle {v} arrives at Node {node} with load {vehicle_load}.")
                vehicle_load -= a[node]  # Subtract the demand at the node
                print(f"    After servicing Node {node}, remaining load: {vehicle_load}")
                
else:
    print("No solution found or solver did not reach optimality gap within time limit.")
```
